##Importando biblioteca Pandas e lendo as planilhas do arquivo .xlsx

```python
import pandas as pd

caract_treino = pd.read_excel('dados_cientista_de_dados_envio.xlsx', sheet_name='dbo.caract_treino')
views_treino = pd.read_excel('dados_cientista_de_dados_envio.xlsx', sheet_name='dbo.materias_views_treino')
caract_novas = pd.read_excel('dados_cientista_de_dados_envio.xlsx', sheet_name='dbo.caract_novas')
views_novas = pd.read_excel('dados_cientista_de_dados_envio.xlsx', sheet_name='dbo.materias_views_novas')
```

##Descrição das informações para averiguar se há valores faltantes e visualização do DataFrame "caract_treino"

```python
caract_treino
```

```python
caract_treino.info()
```

##Conversão dos inteiros para strings da coluna "autor" nos DataFrames "caract_treino" e "caract_novas" para possível utilização das variáveis categóricas da coluna para o modelo de classificação

```python
caract_treino['autor'] = caract_treino['autor'].apply(str)
caract_novas['autor'] = caract_novas['autor'].apply(str)
```

##Substituição dos valores em string da coluna "tipo" para números binários para a elaboração do modelo de classificação

```python
tipo_binario = {'Evergreen': 0,
                'Hard News': 1}

caract_treino['tipo'] = caract_treino['tipo'].map(tipo_binario)
caract_treino
```

##Criação e visualização do DataFrame "x", o qual é o DataFrame das variáveis independentes do modelo, e da Series "y", a qual é a Series da variável dependente. 

```python
x = caract_treino[['autor','tag', 'publieditorial', 'possuilinkafiliacao']]
y = caract_treino['tipo']
```

```python
print(x)
```

```python
print(y)
```

##Transformação das variáveis categóricas, as quais estavam em formato string, em números binários para a elaboração do modelo de classificação, criando o DataFrame "x_dummies"

```python
x_dummies = pd.get_dummies(x)
y_dummies = y

x_dummies
```

##Criação de DataFrame com a lista de todos os autores ordenada pelo "autores_treino", através da junção dos autores dos DataFrames "caract_treino" e "caract_novas", para igualar o número de colunas do "x_dummies" e "x_novas_dummies", com o propósito de ser possível a execução do modelo de classificação e ser executado com maior precisão por conta da influência dos dados sobre os autores

```python
autores_treino = caract_treino['autor'].unique().tolist()
autores_treino
```

```python
len(autores_treino)
```

```python
autores_novas = caract_novas['autor'].unique().tolist()
autores_novas
```

```python
len(autores_novas)
```

```python
for k in range(len(autores_novas)):
  if autores_novas[k] not in autores_treino:
    autores_treino.append(autores_novas[k])
    x_dummies["autor_{}".format(autores_novas[k])] = 0

autores_treino = pd.Series(autores_treino).reset_index().drop(columns='index').rename(columns={0:'autor'})
autores_treino       
```

```python
len(autores_treino)
```

##Visualização do planilha "views_treino" e inserção da coluna "tipo" para análise das matérias 

```python
views_treino
```

```python
tipo = []
for x in range(len(views_treino)):
  for y in range(len(caract_treino)):
    if views_treino['codmateria'][x] == caract_treino['codmateria'][y]:
      tipo.append(caract_treino['tipo'][y])
      break

views_treino['tipo'] = tipo
```

```python
views_treino
```

##Agrupamento pela coluna "codmateria", cálculo do desvio padrão de cada matéria, o qual será usado para a elaboração do modelo por conta das características da "Hard News", a qual possui altos picos e baixas promovendo maior desvio padrão, e da "Evergreen", a qual por conta de ter pouca relação com o tempo tem a tendência de ter menor desvio padrão.

```python
views_treino_por_codmateria = views_treino.groupby(by=['codmateria'])
std_views_treino_por_codmateria = views_treino_por_codmateria.std()
std_views_treino_por_codmateria
```

```python
std_views_treino_por_codmateria = std_views_treino_por_codmateria.reset_index()
std_views_treino_por_codmateria['tipo'] = caract_treino['tipo']
std_views_treino_por_codmateria
```

###Renomeação do nome da coluna {'audiencia':'desvio_padrao'}

```python
std_views_treino_por_codmateria = std_views_treino_por_codmateria.rename(columns={'audiencia':'desvio_padrao'})
std_views_treino_por_codmateria
```

###Elaboração da média entre a categoria do desvio-padrão para ser colocada como critério para a elaboração do modelo

```python
desvio_padrao_medio_views = std_views_treino_por_codmateria['desvio_padrao'].mean()
desvio_padrao_medio_views
```

###Classificação do desvio padrão de cada matéria

```python
desvio_padrao_alto = []
for g in range(len(std_views_treino_por_codmateria)):
  if std_views_treino_por_codmateria['desvio_padrao'][g] <= desvio_padrao_medio_views:
    desvio_padrao_alto.append(0)
  else:
    desvio_padrao_alto.append(1)

std_views_treino_por_codmateria['desvio_padrao_alto'] = desvio_padrao_alto
std_views_treino_por_codmateria
```

###Adição da coluna "desvio_padrao_alto" no "x_dummies" para a elaboração do modelo

```python
x_dummies['desvio_padrao_alto'] = std_views_treino_por_codmateria['desvio_padrao_alto']
```

##Reordenamento das colunas do DataFrame "x_dummies" por ordem alfabética para padronização da ordem das colunas com o DataFrame "x_novas_dummies" e consequente realização bem sucedida do modelo

```python
x_dummies = x_dummies.reindex(sorted(x_dummies.columns), axis=1)
x_dummies
```

##Transformação do DataFrame "x_dummies" e da Series "y_dummies" em array para a elaboração do modelo

```python
x = x_dummies.values
y = y_dummies.values
```

```python
print(x)
```

```python
print(y)
```

##Utilização da biblioteca Scikit-Learn para criação do modelo de classificação com o método "DecisionTreeClassifier", realizando a padronização das características com "StandardScaler", fazendo o treinamento e aplicação do modelo com "train_test_split", medindo a acurácia do modelo com accuracy_score e utilização da biblioteca NumPy para tornar estático o modelo através da função SEED

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

import numpy as np

SEED = 1
np.random.seed(SEED)

raw_treino_x, raw_teste_x, y_treino, y_teste = train_test_split(x, y, random_state = SEED, test_size = 0.3)

scaler = StandardScaler()
scaler.fit(raw_treino_x)
x_treino = scaler.transform(raw_treino_x)
x_teste = scaler.transform(raw_teste_x)

modelo = DecisionTreeClassifier()
modelo.fit(x_treino, y_treino)

resultado = modelo.predict(x_teste)

taxa_de_acerto = accuracy_score(y_teste, resultado) * 100
print('A acurácia foi de %.2f%%.' %taxa_de_acerto)
```

##Criação do modelo bobo, através do "DummyClassifier" da biblioteca Scikit-Learn, para averiguação da qualidade do modelo criado 

```python
from sklearn.dummy import DummyClassifier

dummy_mostfrequent = DummyClassifier()
dummy_mostfrequent.fit(x_treino, y_treino)
acuracia = dummy_mostfrequent.score(x_teste, y_teste) * 100

print('A acurácia do dummy_mostfrequent foi de %.2f%%.' %acuracia)
```
